```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name ="distilgpt2"
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForCausalLM.from_pretrained(model_name)
def chat_with_bot(promot):
  input_ids=tokenizer.encode(promot,return_tensors="pt")
  output=model.generate(input_ids,max_new_tokens=100,pad_token_id=tokenizer.eos_token_id)
  response=tokenizer.decode(output[0],skip_special_tokens=True)
  return response
```

```python
from IPython.display import display
import ipywidgets as widgets
from ipywidgets import Layout

input_box = widgets.Text(
    placeholder='پیام خود را بنویسید...',
    description='شما:',
    disabled=False,
    layout=Layout(width='100%')
)
output_area = widgets.Output()

def handele_message(sender):
    user_input = input_box.value
    input_box.value = ''
    with output_area:
        print(f"شما: {user_input}")
        response = chat_with_bot(user_input)
        print(f"چت بات: {response}")
        print("-" * 50)

input_box.on_submit(handele_message)

display(input_box, output_area)
```
